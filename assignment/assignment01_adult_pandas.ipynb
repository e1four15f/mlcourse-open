```python
import pandas as pd
```

```python
data = pd.read_csv('../data/adult.data.csv')
data.head()
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
# Ваш код здесь

print('Мужчин {0}\nЖенщин {1}'.format(
    len(data[data['sex'] == 'Male']),
    len(data[data['sex'] == 'Female'])))
```

```python
# Лучше решение value_counts()

data['sex'].value_counts()
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
# Ваш код здесь

print('Средний возраст женщин',
    round(data[data['sex'] == 'Female']['age'].mean(), 2))
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
# Ваш код здесь

prop = lambda x: round(len(x) / len(data) * 100, 2)
print('Доля граждан Германии', 
    prop(data[data['native-country'] == 'Germany']))
```

```python
# Лучше решение sum() и shape[0]

(data['native-country'] == 'Germany').sum() / data.shape[0] * 100
```

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? **

```python
# Ваш код здесь

print('Средние значения и среднеквадратичные отклонения возраста\n\
Кто получает более 50K: {0:.3} {1:.3}\n\
Кто получает менее 50K: {2:.3} {3:.3}'.format(
    data[data['salary'] == '>50K']['age'].mean(),
    data[data['salary'] == '>50K']['age'].std(),
    data[data['salary'] == '<=50K']['age'].mean(),
    data[data['salary'] == '<=50K']['age'].std()))
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
# Ваш код здесь 

data[(data['education-num'] <= 10) & (data['salary'] == '>50K')].head()
```

```python
# Лучше решение unique()

data[data['salary'] == '>50K']['education'].unique()
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

```python
# Ваш код здесь

g = data.groupby('race')
g['age'].describe()
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
# Ваш код здесь

dataRichMales = data[(data['sex'] == 'Male') & (data['salary'] == '>50K')]

dataMarriedMales = dataRichMales[(dataRichMales['marital-status'] == 'Married-civ-spouse') | 
                   (dataRichMales['marital-status'] == 'Married-spouse-absent') | 
                   (dataRichMales['marital-status'] == 'Married-AF-spouse')]

print('Женатых мужчин {0}\nХолостых мужчин {1}'.format(
    len(dataMarriedMales), len(dataRichMales) - len(dataMarriedMales)))
```

```python
# Лучше решение isin()

data[(data['sex'] == 'Male') &
     (data['marital-status'].isin(['Never-married', 
                                   'Separated', 'Divorced']))]['salary'].value_counts()
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
# Ваш код здесь

print('Максимальное часов часов в неделю', data['hours-per-week'].max())

print('Количествол людей работающих такое количество часов',
      len(data[data['hours-per-week'] == data['hours-per-week'].max()]))

print('Cреди них процент зарабатывающих много',
len(data[(data['hours-per-week'] == data['hours-per-week'].max()) & (data['salary'] == '>50K')]) / 
     len(data[data['hours-per-week'] == data['hours-per-week'].max()] * 100))

```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
# Ваш код здесь

gRich = data[data['salary'] == '>50K'].groupby('native-country')
gRich['hours-per-week'].describe()
```

```python
gPoor = data[data['salary'] == '<=50K'].groupby('native-country')
gPoor['hours-per-week'].describe()
```

```python
# Другой способ решения 
import numpy as np

pd.crosstab(data['native-country'], data['salary'], 
           values=data['hours-per-week'], aggfunc=np.mean).T
```
